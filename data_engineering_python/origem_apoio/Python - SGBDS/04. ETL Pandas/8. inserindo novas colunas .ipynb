```python



import pandas as pd
df = pd.read_csv(r"K:\Python\Arquivos\Origem\arquivos_csv\Clientes.csv", delimiter=',')
df['created_at'] = pd.to_datetime(df['created_at']) # convert para formato data 
df['ano'] = df['created_at'].dt.year                #adiciona coluna ano
df['Mes'] = df['created_at'].dt.month               # Adicionar coluna com o dia
df['dia'] = df['created_at'].dt.day                 # Adicionar coluna com o dia
df['Hora'] = df['created_at'].dt.hour               # adc hora
df['Minuto'] = df['created_at'].dt.minute           # adc minuto
df['Segundo'] = df['created_at'].dt.second          # adc segundo
# Adicionar coluna com o mês em português
meses = ['janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho',
         'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro']
df['Descmes'] = df['created_at'].dt.month.map(lambda x: meses[x-1])
df

```
