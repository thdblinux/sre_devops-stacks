Conexao psycopg2

```python
import psycopg2
import pandas as pd

# Parâmetros de conexão
dbname   = 'python'
user     = 'postgres'
password = '12345'
host     = 'localhost'
port     = '5432' 

# Criar uma conexão
conexao = psycopg2.connect(dbname=dbname,
                        user=user,
                        password=password,
                        host=host,
                        port=port)

query = """
select 
id
,dt_ocorrencia
,uf
,regiao
,classificacao_da_ocorrencia
from public.anac_mapeamento

"""

df=pd.read_sql_query(query,conexao )
df.head(5)
```

Conexao sqlalchemy

```python
from sqlalchemy import create_engine
import pandas as pd
#conexao com banco de dados
dbname   = 'python'
user     = 'postgres'
password = '12345'
host     = 'localhost'
port     = '5432' 

conexao_str = f'postgresql://{user}:{password}@{host}:{port}/{dbname}'
engine = create_engine(conexao_str)



query2 = """
select
id
,dt_ocorrencia
,uf
from public.anac_mapeamento

"""

df2=pd.read_sql_query(query2,engine )
df2.head(15)



```
