```python
'''
orient determina a orientação do formato JSON resultante

"split" (Padrão):

Os dados do DataFrame são divididos em uma lista de registros (linhas).
Cada registro é um dicionário representando uma linha do DataFrame.
Exemplo: {"index": {"0": 1, "1": 2}, "columns": ["A", "B"], "data": [[1, 4], [2, 5]]}.

"records":
Cada registro representa uma linha e é uma lista de dicionários.
Cada dicionário contém os pares chave-valor representando coluna-valor.
Exemplo: [{"index": 0, "A": 1, "B": 4}, {"index": 1, "A": 2, "B": 5}].

"index":
As chaves são os índices das linhas, e os valores são os registros (dicionários).
Exemplo: {"0": {"A": 1, "B": 4}, "1": {"A": 2, "B": 5}}.

"columns":
As chaves são os nomes das colunas, e os valores são os registros (dicionários).
Exemplo: {"A": {"0": 1, "1": 2}, "B": {"0": 4, "1": 5}}.

"values":
Os valores são uma lista de listas representando os valores das células do DataFrame.
Exemplo: [[1, 4], [2, 5]].

"table":
A saída é um objeto JSON com uma chave schema contendo informações sobre o DataFrame (índices, colunas) e uma chave data contendo uma lista de listas com os dados.
Exemplo: {"schema": {"fields": [{"name": "index", "type": "integer"}, {"name": "A", "type": "integer"}, {"name": "B", "type": "integer"}], "primaryKey": ["index"], "pandas_version": "0.20.0"}, "data": [[0, 1, 4], [1, 2, 5]]}.
'''
```
