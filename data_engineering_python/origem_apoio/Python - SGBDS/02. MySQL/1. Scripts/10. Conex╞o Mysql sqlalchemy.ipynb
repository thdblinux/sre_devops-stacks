```python
from sqlalchemy import create_engine

#Conexão  (Com erro ver abaixo)
eng = create_engine("mysql+pymysql://root:12345@localhost/python")
conn = eng.connect()


query = """  
    SELECT 
        `c`.`name` AS `category`, SUM(`p`.`amount`) AS `total_sales`
    FROM
        (((((`sakila`.`payment` `p`
        JOIN `sakila`.`rental` `r` ON ((`p`.`rental_id` = `r`.`rental_id`)))
        JOIN `sakila`.`inventory` `i` ON ((`r`.`inventory_id` = `i`.`inventory_id`)))
        JOIN `sakila`.`film` `f` ON ((`i`.`film_id` = `f`.`film_id`)))
        JOIN `sakila`.`film_category` `fc` ON ((`f`.`film_id` = `fc`.`film_id`)))
        JOIN `sakila`.`category` `c` ON ((`fc`.`category_id` = `c`.`category_id`)))
    GROUP BY `c`.`name`
    ORDER BY `total_sales` DESC
"""
resultado = conn.execute(query)

# ver linhas
for row in resultado:
    print(row)

conn.close()
eng.dispose() #Libera todos os recursos associados ao objeto eng (o mecanismo SQLAlchemy), incluindo todas as conexões abertas e quaisquer outros recursos mantidos por ele
```

```python
from sqlalchemy import create_engine,text

#Conexão
eng = create_engine("mysql+pymysql://root:12345@localhost/python")
conn = eng.connect()

query = text("""  
    SELECT 
        `c`.`name` AS `category`, SUM(`p`.`amount`) AS `total_sales`
    FROM
        (((((`sakila`.`payment` `p`
        JOIN `sakila`.`rental` `r` ON ((`p`.`rental_id` = `r`.`rental_id`)))
        JOIN `sakila`.`inventory` `i` ON ((`r`.`inventory_id` = `i`.`inventory_id`)))
        JOIN `sakila`.`film` `f` ON ((`i`.`film_id` = `f`.`film_id`)))
        JOIN `sakila`.`film_category` `fc` ON ((`f`.`film_id` = `fc`.`film_id`)))
        JOIN `sakila`.`category` `c` ON ((`fc`.`category_id` = `c`.`category_id`)))
    GROUP BY `c`.`name`
    ORDER BY `total_sales` DESC
""")
resultado = conn.execute(query)

# ver linhas
for row in resultado:
    print(row)

conn.close()
eng.dispose() 
```
